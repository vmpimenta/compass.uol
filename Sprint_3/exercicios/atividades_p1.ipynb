# Exerccio 4

Escreva um cdigo Python para imprimir todos os nmeros primos entre 1 at 100. Lembre-se que voc dever desenvolver o clculo que identifica se um nmero  primo ou no.

Importante: Aplique a funo range().



```python
def primo(n):
    for d in range(2,n):
        if n % d == 0:
            return False
    if n != 1:
        return True

for num in range(1,101):
    if primo(num):
        print(num)
```
