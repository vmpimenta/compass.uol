# Exerccio 21

Implemente duas classes, Pato e Pardal , que herdam de uma superclasse chamada Passaro as habilidades de voar e emitir som.

Contudo, tanto Pato quanto Pardal devem emitir sons diferentes (de maneira escrita) no console, conforme o modelo a seguir.

Imprima no console exatamente assim:

Pato
Voando...
Pato emitindo som...
Quack Quack
Pardal
Voando...
Pardal emitindo som...
Piu Piu



```python
class Passaro:
    def __init__(self,nome, voar, som):
        self.voar = voar
        self.som = som
        self.nome = nome

    def voando(self):
        if self.voar == True:
            return 'Voando...'
        
    def emitir_som(self):
        print(self.nome,'Emitindo som...')
        return self.som
    

Pato = Passaro('Pato', True, 'Quack Quack')
print(Pato.nome)
print(Pato.voando())
print(Pato.emitir_som())

Pardal = Passaro('Pardal', True, 'Piu Piu')
print(Pardal.nome)
print(Pardal.voando())
print(Pardal.emitir_som())
```

# Exerccio 22

Crie uma classe chamada Pessoa, com um atributo privado chamado nome (declarado internamente na classe como __nome) e um atributo pblico de nome id.

Adicione dois mtodos  classe, sendo um para definir o valor de __nome e outro para retornar o valor do respectivo atributo.

Lembre-se que o acesso ao atributo privado deve ocorrer somente atravs dos mtodos definidos, nunca diretamente.  Voc pode alcanar este comportamento atravs do recurso de properties do Python.

Veja um exemplo de como seu atributo privado pode ser lido e escrito:

pessoa = Pessoa(0) 
pessoa.nome = 'Fulano De Tal'
print(pessoa.nome)



```python
class Pessoa: 
    def __init__(self,id):
        self.id = id
        self.__nome = None
    
    def definir_nome(self, nome):
        self.__nome = nome           

    def retornar_nome(self):
        return self.__nome
    
    nome = property(fget=retornar_nome, fset=definir_nome)

pessoa = Pessoa(0)
pessoa.nome = 'Fulano de Tal'
print(pessoa.nome)
```

# Exerccio 23

Crie uma classe  Calculo  que contenha um mtodo que aceita dois parmetros, X e Y, e retorne a soma dos dois. Nessa mesma classe, implemente um mtodo de subtrao, que aceita dois parmetros, X e Y, e retorne a subtrao dos dois (resultados negativos so permitidos).

Utilize os valores abaixo para testar seu exerccio:
x = 4 
y = 5

imprima:
Somando: 4+5 = 9
Subtraindo: 4-5 = -1



```python
class Calculo:
    def __init__(self,x,y):
        self.x = x
        self.y = y
        

    def adicao(self):
        self.resultado = self.x + self.y 
        return 'Somando: {0} + {1} = {2}'.format(self.x, self.y, self.resultado)
        


    def subtracao(self):
        self.resultado = self.x - self.y 
        return 'Subtraindo: {0} + {1} = {2}'.format(self.x, self.y, self.resultado)
        


soma = Calculo(4,5)
print(soma.adicao())
subtrair = Calculo(4,5)
print(subtrair.subtracao())
```

# Exerccio 24

Crie uma classe Ordenadora que contenha um atributo listaBaguncada e que contenha os mtodos ordenacaoCrescente e ordenacaoDecrescente.

Instancie um objeto chamado crescente dessa classe Ordenadora que tenha como listaBaguncada a lista [3,4,2,1,5] e instancie um outro objeto, decrescente dessa mesma classe com uma outra listaBaguncada sendo [9,7,6,8].

Para o primeiro objeto citado, use o mtodo ordenacaoCrescente e para o segundo objeto, use o mtodo ordenacaoDecrescente.

Imprima o resultado da ordenao crescente e da ordenao decresce

[1, 2, 3, 4, 5] [9, 8, 7, 6]



```python
class Ordenadora:
    def __init__(self,listaBaguncada):
        self.listaBaguncada = listaBaguncada

    def ordenacaoCrescente(self):
        self.listaBaguncada = sorted(self.listaBaguncada)
        return self.listaBaguncada

    def ordenacaoDecrescente(self):
        self.listaBaguncada = sorted(self.listaBaguncada, reverse=True)
        return self.listaBaguncada

crescente = Ordenadora([3,4,2,1,5])
print(crescente.ordenacaoCrescente())
descrescente = Ordenadora([9,7,6,8])
print(descrescente.ordenacaoDecrescente())
```
